```python
import pandas as pd
import numpy as np

df = pd.read_csv("Coffee_Chain_Sales .csv")
df.head()
```

```python
print("\nData Info:")
print(df.info())
```

```python
important_columns = ['Sales', 'Profit', 'Marketing_costs']
existing_columns = [col for col in important_columns if col in df.columns]
#finding which columns actually exisit
print("\nExisting Important Columns:", existing_columns)
```

```python
#drop rows with missing values in the exisitng column
if existing_columns:
    df.dropna(subset=existing_columns, inplace=True)
df.head()
```

```python
#converting date column to datatime

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.head()
```

```python
#droping the duplicates
df.drop_duplicates(inplace=True)
```

```python
#creating useful columns
if 'Date' in df.columns:
    df['Month'] = df['Date'].dt.month
df.head()


```

```python
#creating profit margin column
if 'Sales' in df.columns and 'Profit' in df.columns:
    df['Profit_Margin'] = np.where(df['Sales'] != 0, df['Profit'] / df['Sales'], 0)



df.head()



```

```python
#final check of the cleaned data
print("\nFinal data types:")
print(df.dtypes)
```

```python
df.to_csv('Cleaned_Coffee_Chain_Sales.csv', index=False)
print("\nCleaned file saved as 'Cleaned_Coffee_Chain_Sales.csv'")
```
